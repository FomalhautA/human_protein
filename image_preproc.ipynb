```python
import numpy as np

import os

import matplotlib.pyplot as plt
from PIL import Image
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
def shrink_image(source_folder, target_folder):
    """
    Resize images to 256*256 resolution from source folder to target folder.
    """
    fname_lst = os.listdir(source_folder)
    for fname in fname_lst:
        img = Image.open(os.path.join(source_folder, fname))
        
        img2 = img.resize((256, 256), Image.ANTIALIAS)
        img2.save(os.path.join(target_folder, fname))
```

```python
folder = "./Data/train_s"
train_f_lst = os.listdir(folder)
```

```python
def show_image(idx, img_lst):
    filename = img_lst[idx]
    print(filename)
    im = Image.open(os.path.join(folder,filename), 'r')
    
    img_arr = np.array(im.convert('L'))
    print(img_arr.shape)
    
    fig, ax = plt.subplots(1, 2, figsize=(10,10))
    ax = plt.subplot(1, 2, 1)
    ax.imshow(np.asarray(im))
    ax = plt.subplot(1, 2, 2)
    ax.imshow(Image.fromarray(img_arr))
```

```python
show_image(201, train_f_lst)
```
